## PDF: Reading and manipulation

```python
# reading and editing PDF (44)
# pdf are binary file which makes it harder to read

# can extract text and return as string, (images and others cannot)
import PyPDF2

# open as rb(read binary) because PDF is a binary file ( instead of text)
pdfFile = open('meetingminutes1.pdf','rb')

#read pdf
reader= PyPDF2.PdfFileReader(pdfFile)

#number of pages
reader.numPages

# extract text from page 0
page = reader.getPage(0)
page.extractText()


# loop to select all text from all pages
for pageNum in range(reader.numPages):
    print(reader.getPage(pageNum).extractText())
    
```

```python
# TO DO: 'page level' editing, combining 2 pdfs
# basically we can combine multiple pdf using code

import PyPDF2

# open pdf files
pdf1file = open('meetingminutes1.pdf','rb')
pdf2file = open('meetingminutes2.pdf','rb')

# read pdf files
reader1 =PyPDF2.PdfFileReader(pdf1file)
reader2 =PyPDF2.PdfFileReader(pdf2file)
writer = PyPDF2.PdfFileWriter()

# append pdf1 and then pdf2 together

for pageNum in range(reader1.numPages):
    page = reader1.getPage(pageNum)
    writer.addPage(page)
    
for pageNum in range(reader2.numPages):
    page = reader2.getPage(pageNum)
    writer.addPage(page)
    
# save output file
    
outputFile = open('combinedminutes.pdf','wb')
writer.write(outputFile)

#close all opened folders
outputFile.close()
pdf1file.close()
pdf2file.close()
```

## Word doc: Reading and editing

```python
# reading and editing word cocuments (45)

import docx
# document object 
d = docx.Document('demo.docx')
# parargraoh object
d.paragraphs
# string of text in the paragraph
d.paragraphs[0].text

# runs refers to the number of text style in the sentence, 
#in this case there is 4
p = d.paragraphs[1]
p.runs[0].text

# adding text, and formating
p.runs[3].underline = True
p.runs[3].text ='italic and underlined'

# save file
d.save('demo2.docx')



```

```python
# parsing the text from doc, and combining it into a string

import docx 

def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```
