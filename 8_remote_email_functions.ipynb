```python
# Section 15: Email
# sending Emails (46)

import smtplib
conn = smtplib.SMTP('smtp.gmail.com', 587) # domain name and port number, ( can google for this)
type(conn)

# start the connnection
conn.ehlo() # 250 means okya

# start encryption
conn.starttls() # 220 is okay

```

```python
# email and password
# given that most email provides block access, its good to generate app specific password. 
password = 'enterpasswordhere'
conn.login('youremai1@gmail.com', password)
```

```python
# there is a specific format to follow

conn.sendmail('sender@gmail.com', 'receiver@gmail.com', 'subject: So long....\n\nDear Al,\n \
              So long, and thanks for all the fish.\n\n AL')
```

```python
# quit connection
conn.quit
```

```python
# checking email inbox (47)

# extract email from client
import imapclient
#scrap email
import pyzmail

conn = imapclient.IMAPClient('imap.gmail.com', ssl = True)
conn.login('youremail@gmail.com', password)

```

```python
conn.select_folder('INBOX', readonly=True)
```

```python
# all the emails i received, use IMAP search keys (google it)
UIDs = conn.search(b'SINCE 03-Apr-2019')
```

```python
# lets say we want to read 18077, body and flags are like fixed retrieval parameters

rawMessage = conn.fetch(['18077'], ['BODY[]', 'FLAGS'])
message = pyzmail.PyzMessage.factory(rawMessage[18077][b'BODY[]'])
```

```python
# read the email
message.get_subject()
message.get_address('from')
```

```python
message.text_part
```

```python
# to get the message
message.text_part.get_payload().decode('UTF-8')
```
