## 9_GUI automation (simple RPA)

```python
# a more powerful selenium extension, with abit of RPA
# you can move your pointer to any pixel fo the screen
# module is able to detect pointer target by image detection too

# Module: pyautogui
```

```python
# controlling the mouse from python (48)
# FORCE QUIT: you can force your mouse to the top left corner co-ordinate(0,0)

import pyautogui

pyautogui.size()
#size of the screen pixels
width, height = pyautogui.size()
print(width, height)

```

```python
# the coordinates where your mouse is ( X is left-right, Y is up-down)
# starts from top left corner
# you use as a marker to point where u want your mouse to be
pyautogui.position()
```

```python
# computer move the mouse, the duration is to delay stimulate human
pyautogui.moveTo(10,10, duration=1.5)

# move to the right
pyautogui.moveRel(200,0, duration =1.5)

#move up the screen ( negative)
pyautogui.moveRel(0, -100, duration = 1.5)
```

```python
#click on the pointer
pyautogui.click(272,1096)
```

```python
# controlling keyboard from python (49)

# move cursor and click and type
pyautogui.click(578,533); pyautogui.typewrite('Hello World', interval = 0.2)

# this include keyboard movement keys like 'left'
pyautogui.click(578,533); pyautogui.typewrite(['a','b','hello','left','left','Z'], interval = 0.2)
```

```python
# list of keyboard functions that is available with pyautogui
pyautogui.KEYBOARD_KEYS
```

```python
# press single key

pyautogui.press('f1')
```

```python
# hot keys ctrl + o

pyautogui.hotkey('ctrl','o')
```

```python
# screenshots and image recognition (50)

import pyautogui
# takes a screenshot, optional save it in a fike
pyautogui.screenshot('screenshot.png')
```

```python
#locate on screen function, 
# feed in an image and the computer will be able to locate it on your screen
# returns the coordinates

# will only wor if its exact match
pyautogui.locateOnScreen('button7.PNG')
pyautogui.locateCenterOnScreen('button7.PNG')
```
